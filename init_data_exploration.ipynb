```python
pd.set_option("max_colwidth", 400)
offer_completed_split = offer_completed['value'].str.split("'",expand=True)
offer_completed['value'] = offer_completed_split[3]
offer_completed
```

```python
offer_completed_counts = offer_completed.groupby(['person','value'], as_index=False).count()
offer_completed_counts = offer_completed_counts[["person", "value", "event"]].rename(columns={"event":"per_offer_type"})
offer_completed_counts

#TIMES EACH PERSON COMPLETED EACH OFFER TYPE
```

```python
offer_completed_counts = offer_completed.groupby(['person'], as_index=False).count()
offer_completed_counts = offer_completed_counts[["person", "event"]].rename(columns={"event":"per_person"})
offer_completed_counts.sort_values('per_person')

#TIMES EACH PERSON COMPLETED A TRANSACTION
```

```python
transaction_split = transaction['value'].str.split(": ",expand=True)
transaction_split = transaction_split[1].str.split("}",expand=True)
transaction['value'] = transaction_split[0].astype(float)
transaction
```

```python
total_spent = transaction.groupby(['person'], as_index=False)['value'].sum()
total_spent
total_spent.sort_values('value')
#TOTAL AMOUNT SPENT IN TIMEFRAME PER PERSON
```
