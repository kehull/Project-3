```python
age_encode=df['age']
label_encoder.fit(age_encode)
encoded_age=label_encoder.transform(age_encode)
df['age']=encoded_age
```

```python
X_2=df[['customer_id','event','time','gender','offer_id/transaction_amount','income']]
y_2=df['gender']
```

```python
label_encoder.fit(y_2)
encoded_y_2=label_encoder.transform(y_2)
y_2=encoded_y_2.reshape(-1,1)
```

```python
X_2_train,X_2_test,y_2_train,y_2_test=train_test_split(X_2,y_2,random_state=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
X_2_scaler = MinMaxScaler().fit(X_2_train)
X_2_train_scaled = X_2_scaler.transform(X_2_train)
X_2_test_scaled = X_2_scaler.transform(X_2_test)
```

```python
y_2_train_categorical = to_categorical(y_2_train)
y_2_test_categorical = to_categorical(y_2_test)
```

```python
model_2 = Sequential()
model_2.add(Dense(units=300, activation='relu', input_dim=X_2.shape[1]))
model_2.add(Dense(units=300, activation='relu'))
model_2.add(Dense(units=300, activation='relu'))
model_2.add(Dense(units=3, activation='softmax'))
```

```python
model_2.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
monitor_2 = EarlyStopping(monitor='accuracy', min_delta=1e-3, patience=5, 
        verbose=2, mode='auto', restore_best_weights=True)
model_2.fit(
    X_2_train_scaled,
    y_2_train_categorical,
    callbacks=[monitor_2],
    epochs=100,
    shuffle=True,
    verbose=2
) 
```
