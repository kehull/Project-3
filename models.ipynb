```python
#import dependencies
import pandas as pd
import numpy as np
from numpy.random import seed
seed(1)
```

```python
#read in csv
df=pd.read_csv("Resources/machine_learning_data.csv")
```

```python
#preview data
df.head()
```

```python
#drop unused column(s)
df.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
#check data types
df.dtypes
```

```python
#filter df to offer completed
df=df.loc[df['event']== 'offer completed']
```

```python
#display unique offers for reference in the flask app
df['offer_id/transaction_amount'].unique()
```

```python
#encode data
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
customer_encode=df['customer_id']
label_encoder.fit(customer_encode)
encoded_customer=label_encoder.transform(customer_encode)
df['customer_id']=encoded_customer
```

```python
#reset inded
df.reset_index(inplace=True)
```

```python
event_encode=df['offer_id/transaction_amount']
label_encoder.fit(event_encode)
encoded_event=label_encoder.transform(event_encode)
df['offer_id/transaction_amount']=encoded_event
```

```python
gender_encode=df['gender']
label_encoder.fit(gender_encode)
encoded_gender=label_encoder.transform(gender_encode)
df['gender']=encoded_gender
```

```python
#drop extra index column
df.drop(columns=['index'],inplace=True)
```

```python
time_encode=df['time']
label_encoder.fit(time_encode)
encoded_time=label_encoder.transform(time_encode)
df['time']=encoded_time
```

```python
event_encode=df['event']
label_encoder.fit(event_encode)
encoded_event=label_encoder.transform(event_encode)
df['event']=encoded_event
```

```python
#postview of the data
df
```

```python
#prepare data for split
X=df[['age','income']]
y=df['offer_id/transaction_amount']
# alternative
# X=df[['customer_id','event','time','gender','age','income']]
# y=df['offer_id/transaction_amount']

```

```python
#encode y data
label_encoder.fit(y)
encoded_y=label_encoder.transform(y)
np.unique(encoded_y)
```

```python
#reshape y
y=encoded_y.reshape(-1,1)
print(X.shape,y.shape)
```

```python
#split data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=1)
```

```python
#scale data
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y_minmax = MinMaxScaler().fit(y_train)
y_train_minmax = y_minmax.transform(y_train)
y_test_minmax = y_minmax.transform(y_test)
```

```python
#one hot encode y data
from tensorflow.keras.utils import to_categorical
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
```

```python
#build random forest model
from sklearn.ensemble import RandomForestClassifier
model_2= RandomForestClassifier(class_weight='balanced',criterion='entropy',max_features='log2',
                                max_samples=101, n_estimators=700)
model_2.fit(X_train_scaled,y_train.ravel())


#print scores of training and test data
print(f"Training Data Score: {model_2.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model_2.score(X_test_scaled, y_test)}")
```

```python
#form a grid search to determine best parameters
from sklearn.model_selection import GridSearchCV
param_grid = {'n_estimators': list(range(1,1000,100)),
             'max_samples':list(range(1,1000,100)),
             'criterion':['gini','entropy'],
             'max_features':['auto','sqrt','log2'],
             'class_weight':['balanced', 'balanced_subsample']}
              
grid = GridSearchCV(model_2, param_grid, verbose=3)
```

```python
#build grid
grid.fit(X_train_scaled,y_train.ravel())
```

```python
#check best grid params and scores
print(grid.best_params_)
print(grid.best_score_)
```

```python
#save model
import joblib
filename = 'models/random_forest_model.h5'
joblib.dump(model_2, filename)
```
