```python
#build random forest model
from sklearn.ensemble import RandomForestClassifier
model_2= RandomForestClassifier(class_weight='balanced',criterion='entropy',max_features='log2',
                                max_samples=101, n_estimators=700)
model_2.fit(X_train_scaled,y_train.ravel())


#print scores of training and test data
print(f"Training Data Score: {model_2.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model_2.score(X_test_scaled, y_test)}")
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'n_estimators': list(range(1,1000,100)),
             'max_samples':list(range(1,1000,100)),
             'criterion':['gini','entropy'],
             'max_features':['auto','sqrt','log2'],
             'class_weight':['balanced', 'balanced_subsample']}
              
grid = GridSearchCV(model_2, param_grid, verbose=3)
```

```python
grid.fit(X_train_scaled,y_train.ravel())
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
