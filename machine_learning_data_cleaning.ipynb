```python
#import dependencies
import pandas as pd
```

```python
#read in csv's
portfolio_df=pd.read_csv("Resources/portfolio.csv",index_col=0)
profile_df=pd.read_csv("Resources/profile.csv",index_col=0)
transcript_df=pd.read_csv("Resources/transcript.csv",index_col=0)
```

```python
#drop na
portfolio_df.dropna(inplace=True)
profile_df.dropna(inplace=True)
transcript_df.dropna(inplace=True)
```

```python
#merge df's
combined_customer_data=transcript_df.merge(profile_df,how='right',left_on='person',right_on='id')
```

```python
combined_customer_data.drop(labels='id', axis=1, inplace=True)
```

```python
#rename columns
combined_customer_data.rename(columns={'person':'customer_id'},inplace=True)
```

```python
combined_customer_data
```

```python
#check dtypes
combined_customer_data.dtypes
```

```python
#split out value
combined_customer_data['offer_id/transaction_amount']=combined_customer_data['value'].str.split(":",expand=True)[1]
```

```python
#drop extra column
combined_customer_data.drop(columns=['value'],inplace=True)
```

```python
combined_customer_data['offer_id/transaction_amount'] = combined_customer_data['offer_id/transaction_amount'].str.replace('}', '')
```

```python
#remove "{" from column and reorder data
combined_customer_data=combined_customer_data[['customer_id','event','offer_id/transaction_amount','time','gender',
                                               'age','income','became_member_on']]
combined_customer_data
```

```python
#save df to resources
combined_customer_data.to_csv("Resources/machine_learning_data.csv")
```
